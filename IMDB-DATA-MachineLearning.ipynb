# K-Means Clustering

# Clustering the dataset according to METASCORE

```python

# clustering dataset according to metascore 
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
 
x1 = np.array(z) #RATING
x2 = np.array(y) #METASCORE 
 
plt.plot()
plt.xlim([0,10])
plt.ylim([0,125])
plt.title('Dataset')
plt.scatter(x1,x2)
plt.show()
```

```python
# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']
 
# KMeans algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)
 
plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 100])
 
plt.show()
 
```

```python
# clustering dataset
# determine k using elbow method
 
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
 
  
x1 = np.array(y) #RATING
x2 = np.array(m) #METASCORE

 
plt.plot()
plt.xlim([0, 100])
plt.ylim([0, 600])
plt.title('Dataset')
plt.scatter(x1, x2)
plt.show()
 
# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']
 
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
 
```

# NEURAL NETWORK

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Train the scaler, which standarizes all the features to have mean=0 and unit variance
sc = StandardScaler()
sc.fit(train_X)
```

```python
# Apply the scaler to the X training data
X_train_std = sc.transform(train_X)
print("Train")
print(X_train_std)
# Apply the SAME scaler to the X test data
print("Test")
X_test_std = sc.transform(test_X)
print(X_test_std)
```

```python
# Create a perceptron object with the parameters: 40 iterations (epochs) over the data, and a learning rate of 0.1
ppn = Perceptron(max_iter=40, eta0=0.1, random_state=0)

# Train the perceptron
ppn.fit(X_train_std, train_y)
```

```python
# Apply the trained perceptron on the X data to make predicts for the y test data
y_pred = ppn.predict(X_test_std)
print(X_test_std)
print(y_pred) #these are the predicted metascores
print(mean_squared_error(test_y, y_pred))
```

```python
print('Accuracy: %.2f' % accuracy_score(test_y, y_pred))
```
